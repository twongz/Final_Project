```python
heatmap_df = diamonds
heatmap_df.drop('Unnamed: 0', axis=1, inplace=True)
```

In the visualization below, I deleted a column that was titled "Unnamed".

```python
plt.subplots(figsize=(20,15))
corr = heatmap_df.corr()
sns.heatmap(corr)

tw_map = sns.heatmap(data=corr, square=True, annot=True, cbar=False)
```

For my final visualization, I created a heatmap to visualize the data in the data frame. I used the "square=True" attribute to have the results appear in every grid. Because the sizes of the squares in the heatmap are going to be identical, I set "annot=True" so it would be easier to compare the visuals with the gradient on the right. 

The lowest frequencies are represented by the darker colors and the higher frequencies are represented by the lighter colors.

From the map, price is most correlated to carats. The second highest factors are x,y, and z or length, width, and depth.
